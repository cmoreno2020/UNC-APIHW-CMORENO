```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import pprint
import json

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
ncities = len(cities)
print (ncities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Save config information
url = "http://api.openweathermap.org/data/2.5/weather?"

i = 1

cdata = []

#Iterate in cities, and get data from OpenWeather API - store data in cdata
for city in cities:
    
    
    query_url = url + "appid=" + api_key + "&q=" + city

    print ("Processing city: ",i," of ",ncities," | ",city)
    i += 1

    # Get weather data
    try: 
        weather_response = requests.get(query_url)
        weather_json = weather_response.json()
        cdata.append(weather_json)
    except:
        pass

    
print ("END *****************")    
    

# Get the temperature from the response
#pprint.pprint(f"The weather API responded with: {weather_json}.")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#Save data from API into a file (data1.txt)
with open('data1.txt', 'w') as outfile:  
    json.dump(cdata, outfile)
 
print (len(cdata))
```

```python
#For testing purposes - read back text file
with open('data1.txt') as json_file:  
    cdatafile = json.load(json_file)
pprint.pprint(len(cdatafile))
```

```python
#Create lists with required information for each city read
cityn = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
MTemp = []
WSpeed = []

for i in range(0,len(cdatafile)):
    try:
        cityn.append(cdatafile[i]["name"])
        cloudiness.append(cdatafile[i]['clouds']['all'])
        country.append(cdatafile[i]['sys']['country'])
        date.append(cdatafile[i]['dt'])
        humidity.append(cdatafile[i]['main']['humidity'])
        lat.append(cdatafile[i]['coord']['lat'])
        lng.append(cdatafile[i]['coord']['lon'])
        kelvin = cdatafile[i]['main']['temp_max']
        F = ((kelvin - 273.15)* (9/5)) + 32
        MTemp.append(F)
        WSpeed.append(cdatafile[i]['wind']['speed']*0.000621371 * 3600)
    except:
        pass
    
#Formula for converting Kelvin to F: (0K − 273.15) × 9/5 + 32 = -459.7°F
#Formula to convert m/s to Miles/Hour: (m/s)* 0.000621371 * 3600
```

```python
#Create a data frame with required information for the cities
df1 = pd.DataFrame({"City": cityn,
                 "Cloudiness" : cloudiness,
                 "Country" : country,
                 "Date": date,
                 "Humidity": humidity,
                 "Lat" : lat,
                 "lng" : lng,
                 "Max Temperature": MTemp,
                 "Wind Speed" : WSpeed})
```
